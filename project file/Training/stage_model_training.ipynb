```python
import pandas as pd
import os

# Absolute path to the Data folder
data_path = r"C:\Users\manid\OneDrive - Seshadri Rao Gudlavalleru Engineering College\Desktop\liver_chirosis\Data"

# Show what's inside the folder
print(os.listdir(data_path))

# Load the CSV
df = pd.read_csv(os.path.join(data_path, "liver_cirrhosis.csv"))

# Show first few rows
df.head()
# 📊 Performance Testing Code for Liver Disease Prediction

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pickle

# Load the data
X_train = pd.read_csv("C:/Users/manid/OneDrive - Seshadri Rao Gudlavalleru Engineering College/Desktop/liver_chirosis/Data/X_train.csv")
y_train = pd.read_csv("C:/Users/manid/OneDrive - Seshadri Rao Gudlavalleru Engineering College/Desktop/liver_chirosis/Data/y_train.csv")


# Load the model
with open("C:/Users/manid/OneDrive - Seshadri Rao Gudlavalleru Engineering College/Desktop/liver_chirosis/Flask/rf_stage.pkl", "rb") as f:
    model = pickle.load(f)

# ✅ 1. Model Summary - Feature Importance
feature_names = X_train.columns
importances = model.feature_importances_
indices = np.argsort(importances)

plt.figure(figsize=(10, 6))
sns.barplot(x=importances[indices], y=feature_names[indices])
plt.title("Feature Importances")
plt.xlabel("Importance")
plt.ylabel("Features")
plt.tight_layout()
plt.show()

# ✅ 2. Accuracy (Training & Validation)
from sklearn.model_selection import train_test_split

X_train_split, X_test_split, y_train_split, y_test_split = train_test_split(
    X_train, y_train, test_size=0.2, random_state=42
)

train_acc = model.score(X_train_split, y_train_split)
val_acc = model.score(X_test_split, y_test_split)

print("Training Accuracy:", round(train_acc * 100, 2), "%")
print("Validation Accuracy:", round(val_acc * 100, 2), "%")

# ✅ 3. (Optional) Fine Tuning - If GridSearchCV was done
# from sklearn.model_selection import GridSearchCV
# param_grid = {
#     'n_estimators': [100, 200],
#     'max_depth': [None, 10, 20]
# }
# grid = GridSearchCV(model, param_grid, cv=5)
# grid.fit(X_train_split, y_train_split)
# 
# print("Best Parameters:", grid.best_params_)
# print("Best Validation Score:", grid.best_score_)


```
